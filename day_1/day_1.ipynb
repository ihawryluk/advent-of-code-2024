# Part one

```python
def sort_columns(df):
    sorted_col1 = df['col1'].sort_values().reset_index(drop=True)
    sorted_col2 = df['col2'].sort_values().reset_index(drop=True)
    sorted_df = pd.DataFrame({'col1': sorted_col1, 'col2': sorted_col2})
    return sorted_df

def add_distance(df):
    df['distance'] = np.abs(df['col1'] - df['col2'])
    return df

def get_total_distance(df):
    df = sort_columns(df)
    df = add_distance(df)
    return df['distance'].sum()
```

```python
input_test = """
3   4
4   3
2   5
1   3
3   9
3   3
"""

data = StringIO(input_test)

# Read the data into a DataFrame
input_test_df = pd.read_csv(data, sep=r'\s+', header=None, names=['col1', 'col2'])
input_test_df
```

# Part Two

```python
# for each number from the list on the left, we want number of instances in the column on the right

def find_matching_numbers(df, number):
    number_of_matches = (df['col2'] == number).sum()
    return number_of_matches

def get_single_similarity(df, number):
    return number * find_matching_numbers(df, number)

def get_total_similarity(df):
    total = 0
    for number in df['col1'].values:
        total += get_single_similarity(df, number)
    return total
```
